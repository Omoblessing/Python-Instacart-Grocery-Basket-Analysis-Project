```python
# Define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return np.nan
```

```python
# Apply the function
ords_prods_merge['price_range'] = ords_prods_merge.apply(price_label, axis=1)
```

```python
# Creating a new column for busiest and slowest days
result = []

for value in ords_prods_merge["order_dow"]:
    if value == 0 or value == 1:
        result.append("Busiest days")
    elif value == 3 or value == 4:
        result.append("Slowest days")
    else:
        result.append("Regularly busy")
```

```python
#  Creating new column busiest_period_of_day to identify time periods “Most orders,” “Average orders,” and “Fewest orders.”"
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# For the following orders: "Fewest orders" = 0,1,2,3,4,5,6,7,20,21,22,23; "Most orders" = 10,11,12,13,14,15,16; "Average orders" = 8,17,18,19
```

```python
# Creating a new column for Most orders, average orders,and fewest orders
result = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value in [10,11,12,13,14,15,16,9]:
        result.append("Most orders")
    elif value in [0,1,2,3,4,5,6,7,20,21,22,23]:
        result.append("Fewest orders")
    else:
        result.append("Average orders")
```

```python
result
```

```python
# creating a new column within your ords_prods_merge dataframe and set it equal to result
ords_prods_merge ['busiest_period_of_day'] = result
```
